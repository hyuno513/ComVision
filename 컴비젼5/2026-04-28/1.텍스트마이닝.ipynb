```python
# 텍스트마이닝_형태소분석

# 토그나이징 : 자연어 처리를 위해서는 우선 텍스트에 대한 정보를 단위별로 나누는 것이 일반적
# 예를 들어, 영화 리뷰 내용을 예측한다고 하면 한 문장을 단어 단위로 쪼개서 분석할 수 있음
# 이처럼 예측해야 할 입력 정보(문장 또는 발화)를 하나의 특정 기본 단위로 자르는 것을 트그나이징
# 파이썬을 이용하면 이러한 작업을 라이브러리를 통해 간편하게 처리
# 일정한 의미가 있는 가장 작은 단위로 정보를 나누는 작업. 텍스트에 대해 특정 기준 단위로 문장을 나누는 것을 의미
# 형태소 : 언어학에서 사용하는 요어이며, 일정한 의미가 있는 가장 작은 말의 단위
# 형태소 분석 (Morphological Analysis)은 자연 언어의 문장을 '형태소'라는 의미를 갖는 최소 단위 문장을 나누는 것을 의미

# 영어 형태소 분석은 크게 어렵지 않은데, 형태소 마다 띄어쓰기를 해서 문장을 구성하는 것이 기본이기 때문
# 한글의 경우 형태소 마다 띄어쓰기를 하지 않기 때문에, 품사 사전과 문법 사전을 기반으로 대조하면서 형태소를 분석

# 수많은 형태소 라이브러리가 오픈소스로 배포되고 있음
# 많이 사용중인 라이브러리가 '파이썬 한국어 형태소 분석 라이브러리 KoNLPY'
# 공식 홈페이지 : https://konlpy.org

# KoNLPy는 한글 자연어 처리를 쉽고 간결하게 처리할 수 있도록 만들어진 오픈소스 라이브러리
# 또한 국내에 이미 만들어져 사용되고 있는 여러 형태소 분석기를 사용할 수 있기 허용
# 이를 이용하면 한나눔, 꼬모란, Komoran, MeCab, 트위터 등의 형태소 분석기를 쉽게 이용할 수 있음

# 자바 패키지도 설치가 되어야함
# 1. koNLPy 패키지 설치
# koNLPy 패키지를 이용하면 한글 텍스트로 형태소 분석이 가능

from konlpy.tag import Okt
okt = Okt()
text = '아버지가 방에 들어갑니다.'

# 형태소 추출
# 1. 인자로 입력한 문장을 형태소 단위로 토크나이징 한다. 토크나이징된 형태소들은 리스트 형태로 반환
morphs = okt.morphs(text)
print(morphs)
```

```python
# 형태소와 품사 태그 추출
# pos() : 인자로 입력한 문장에서 형태소를 추출한 뒤 품사 태깅을 한다.
# 추출된 형태소와 그 형태소의 품사가 형태로 묶어서 리스트로 반환된다.
pos = okt.pos(text)
print(pos)
```

```python
# 명사만 추출
# nouns() : 인자로 입력한 문장에서 품사가 명사인 토큰만 추출
nouns = okt.nouns(text)
print(nouns)
```

```python
# 정규화, 어구 추출
# normalize : 입력한 문장을 정규화 시킨다. 오타가 섞인 문장을 정규화 처리하는데 효과적
# phrases : 입력한 문장에서 우구를 추출한다
text = '오늘 날씨가 좋아욬ㅋㅋㅋㅋㅋㅋㅋㅋ'
print(okt.normalize(text))
print(okt.phrases(text))
```
