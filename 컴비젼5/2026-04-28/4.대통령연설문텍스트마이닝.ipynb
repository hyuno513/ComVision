```python
# 가장 많이 사용된 단어 알아보기

# 1. 연설문 불러오기
```

```python
txt_file = open('./input/speech_Roh.txt', encoding='utf-8').read()
print(txt_file)
```

```python
# 02. 불필요한 문자 제거하기
# 텍스트 파일 출력하면 특수문자, 한자. 공백 등이 포함되어 있을 수 있음
# 이런 요소는 분석 대상이 아니므로 제거해야 함
# 문자 처리 패키지인 re의 sub()을 이용해 한글이 아닌 모든 문자를 공백으로 바꿈
```

```python
import re

txt_file_01 = re.sub('[^가-힣]',' ',txt_file)
print(txt_file_01)
```

```python
# 03 명사 추출하기
# 형태소 중에도 명사를 보면 텍스트가 무엇에 관한 내용인지 파악할 수 있기 때문에
# 텍스트에서 명사만 추출해 분석하는 경우가 많음
# konlpy.tag.Hannanum()의 nouns()를 이용하면 문장에서 명사를 추출할 수 있다.

import konlpy

hannanum = konlpy.tag.Hannanum()
print(hannanum.nouns('세계의 안보'))
```

```python
# 연설문에서 명사만 추출하기
nouns = hannanum.nouns(txt_file_01)
print(nouns)
```

```python
import pandas as pd
df_word = pd.DataFrame({'word': nouns})
print(df_word)
```

```python
# 단어 빈도표 만들기
# 먼저 한 글자로 된 단어는 의미가 없는 경우가 많으므로 제거

df_word['count'] = df_word['word'].str.len()
print(df_word)
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >=2')
print(df_word.sort_values('count'))
```

```python
# 단어의 사용 빈도를 구하고 빈도순으로 출력
df_word = (df_word.groupby('word',as_index=False) # 단어별 분리
           .agg(n=('word', 'count')) # 빈도 구하기
           .sort_values('n', ascending=False))
print(df_word)
```

```python
# 단어 빈도 막대 그래프 만들기
top20 = df_word.head(20)
print(top20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.barplot(data=top20, y='word', x='n')
plt.show()

df_word.to_csv('./data/speech_Roh_01.csv', mode='w')
```
