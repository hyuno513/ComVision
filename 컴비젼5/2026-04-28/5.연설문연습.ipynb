```python
txt_file1 = open('./input/speech1.txt', encoding='utf-8').read()
print(txt_file1)
```

```python
import re

txt_file_01 = re.sub('[^가-힣]',' ',txt_file1)
print(txt_file_01)
```

```python
import konlpy

hannanum = konlpy.tag.Hannanum()
nouns = hannanum.nouns(txt_file_01)

import pandas as pd
df_word1 = pd.DataFrame({'word': nouns})
# print(df_word)
df_word1['count'] = df_word1['word'].str.len()

df_word1 = df_word1.query('count >=2')
# print(df_word1.sort_values('count'))
df_word1 = (df_word1.groupby('word',as_index=False) # 단어별 분리
           .agg(n=('word', 'count')) # 빈도 구하기
           .sort_values('n', ascending=False))
print(df_word1)
```

```python
top1 = df_word1.head(20)
print(top1)

import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.barplot(data=top1, y='word', x='n')
plt.show()

df_word.to_csv('./data/speech1.csv', mode='w')
```
