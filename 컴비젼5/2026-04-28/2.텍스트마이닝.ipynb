```python
from konlpy.tag import Kkma

kkma = Kkma()
text = '아버지가 방에 들어갑니다.'

# 리스트 형식
morphs = kkma.morphs(text)
print(morphs)

# 튜플 형태로 묶어서 리스트로 반환
pos = kkma.pos(text)
print(pos)

nouns = kkma.pos(text)
print(nouns)
```

```python
# 문장 분리
# sentences : 인자로 입력한 여러 문장을 분리해주는 역할. 분리된 문장은 리스트 형태로 반환

sentences = "오늘 날씨는 어때요? 내일은 덥다던데"
s = kkma.sentences(sentences)
print(s)
```
