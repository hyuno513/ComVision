```python
# 간단한 딥러닝 모델
# 간단한 구조의 인공 신경망으로 1차 선형회귀식을 찾는 문제
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

print(torch.__version__)

# 모델 학습에 사용할 입력 데이터를 준비
# y = x + 1 관계를 갖는 숫자를 x, y 변수에 각각 10개씩 입력
x = torch.FloatTensor([-3, 31, -11, 4, 0, 22, -2, -5, -25, -14]).view(-1, 1)
y = torch.FloatTensor([-2, 32, -10, 5, 1, 23, -1, -4, -24, -13]).view(-1, 1)
# print(x)
# print(y)
```

```python
# 모델 설정 (Sequential API와 유사한 모델을 사용)
# nn.Linear(입력 차원, 출력 차원) -> 텐서플로우의 Dense(units=1, input_dim=1)와 동일
model = nn.Sequential(
    nn.Linear(1, 1),
)

# 컴파일 단계 (손실 함수와 옵티마이저 설정)
critical = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 모델 학습 (fit)과정
epochs = 3000

for epoch in range(epochs):
    # 예측
    output = model(x)
    loss = critical(output, y)

    # 역전파 및 최적화
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 500 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')

# 결과 확인
print('\n[학습된 가중치 확인]')
for name, param in model.named_parameters():
    print(f'{name} : {param.data}')

# 예측
test_data = torch.FloatTensor([11, 22, 33]).view(-1, 1)
with torch.no_grad(): # 예측 시에는 그래디언트 계산 제외
    prediction = model(test_data)
    print("\n[예측결과]")
    print(prediction.numpy())
```
